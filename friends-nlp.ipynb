```python
# Function to check if the episode exists in the dataset
def check_episode_exists(season_num, episode_num, dataset_folder):
    episode_filename = f"S{season_num:02d}E{episode_num:02d}.txt"
    for file_name in os.listdir(dataset_folder):
        if episode_filename in file_name:
            return os.path.join(dataset_folder, file_name)
    return None

# User input for season and episode number
season_num = int(input("Enter the season number: "))
episode_num = int(input("Enter the episode number: "))

# Adjusted dataset folder after inspecting the directory
dataset_folder = '/kaggle/input/friends-tv-series-screenplay-script'  # Use the correct folder name based on the output

# Check if the episode exists
episode_path = check_episode_exists(season_num, episode_num, dataset_folder)

if episode_path:
    print(f"Episode file found: {episode_path}")
else:
    print(f"Episode S{season_num:02d}E{episode_num:02d} not found in the dataset.")
```
