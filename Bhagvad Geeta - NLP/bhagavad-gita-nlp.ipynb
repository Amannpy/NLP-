```python
def most_similar_shloka(encoder, text, shloka_embeddings, shlokas_list):
    text_vector = encoder.encode([text])
    # find the most similar shloka with distance
    distances = np.linalg.norm(shloka_embeddings - text_vector, axis = 1)
    min_index = np.argmin(distances)
    # return the corresponsing row
    return shlokas_list[min_index]
```
